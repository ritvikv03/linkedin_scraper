```python
# [SQL,r_lang,Python,Tableau,Powerbi,Azure,ML,Google_Cloud,Excel,AWS,Hadoop,MongoDB,NLP,Phd,Spark]
def vectorize(df_merged):
    df_merged['Spark']=df_merged.content.apply(lambda x: Spark(x))
    df_merged['SQL']=df_merged.content.apply(lambda x: SQL(x))
    df_merged['R_lang']=df_merged.content.apply(lambda x: r_lang(x))
    df_merged['Python']=df_merged.content.apply(lambda x: Python(x))
    df_merged['Tableau']=df_merged.content.apply(lambda x: Tableau(x))
    df_merged['Powerbi']=df_merged.content.apply(lambda x: Powerbi(x))
    df_merged['Azure']=df_merged.content.apply(lambda x: Azure(x))
    df_merged['ML']=df_merged.content.apply(lambda x: ML(x))
    df_merged['Excel']=df_merged.content.apply(lambda x: Excel(x))
    df_merged['Google_Cloud']=df_merged.content.apply(lambda x: Google_Cloud(x))
    df_merged['AWS']=df_merged.content.apply(lambda x: AWS(x))
    df_merged['Hadoop']=df_merged.content.apply(lambda x: Hadoop(x))
    df_merged['MongoDB']=df_merged.content.apply(lambda x: MongoDB(x))
    df_merged['NLP']=df_merged.content.apply(lambda x: NLP(x))
    df_merged['Sklearn']=df_merged.content.apply(lambda x: Sklearn(x))
    df_merged['Phd']=df_merged.content.apply(lambda x: Phd(x))
    df_merged['Scala']=df_merged.content.apply(lambda x: Scala(x))
    df_merged['Sagemaker']=df_merged.content.apply(lambda x: Sagemaker(x))
    df_merged['Tensorflow']=df_merged.content.apply(lambda x: Tensor(x))
    df_merged['Julia']=df_merged.content.apply(lambda x: Julia(x))
    return df_merged
```

```python
def convert(df_merged):
    requirements = df_merged.iloc[:,8:]
    apriori_table=apriori(requirements,min_support=0.002,use_colnames=True)
    apriori_table['length'] = apriori(requirements,min_support=0.002,use_colnames=True).itemsets.apply(lambda x:len(x))
    apriori_table['items']= apriori_table.itemsets.astype('str').replace('frozenset','',regex=True).replace('[\(\)\{\}\'\']','',regex=True)
    apriori_table = apriori_table.sort_values('support',ascending=False)
    return apriori_table
```

```python
entry_levels = df_merged[df_merged.level=='Entry level']
entry_level_apriori=convert(vectorize(entry_levels))
entry_level_apriori[entry_level_apriori.length==1]
```

```python
def Support_levels(pair_size = 2,levels='Entry level', employment = 'Full-time'):
    entry_levels = df_merged[df_merged.employment== employment]
    entry_level_apriori=convert(vectorize(entry_levels))
    return entry_level_apriori[entry_level_apriori.length== pair_size].drop(columns=['itemsets'])
Support_levels(1,employment='Full-time').head(15)
```

```python
def reqs(pair_size = 2):
    entry_levels = df_merged
    entry_level_apriori=convert(vectorize(entry_levels))
    return entry_level_apriori[entry_level_apriori.length== pair_size].drop(columns=['itemsets'])
reqs(1)
```

```python
higher_levels = df_merged[(df_merged.level=='Mid-Senior level') |(df_merged.level=='Executive') ]
higher_level_apriori=convert(vectorize(higher_levels))
association =association_rules(higher_level_apriori,metric='lift',min_threshold=1)
```

```python
entry_ = Support_levels(1,levels='Entry level')                 # Get the requirement list for entry level jobs
entry_ = entry_.rename(columns={'support':'Entry'})             # Change a column name for appropriate formatting
assoc = Support_levels(1,levels='Associate')                    # Get the requirement list for associate level jobs
assoc = assoc.rename(columns={'support':'Associate'})           # Change a column name for appropriate formatting
mid_senior = Support_levels(1,levels='Mid-Senior level')        # Get the requirement list for mid-senior level jobs
mid_senior = mid_senior.rename(columns={'support':'Mid-Senior'}) # Change a column name for appropriate formatting
all_levels =  reqs(1)                                           # Get the requirement list for all jobs 
all_levels = all_levels.rename(columns={'support':'All'})       # Change a column name for appropriate formatting
```

```python
all_data = pd.DataFrame()
e = pd.pivot_table(data=entry_,columns='items',values='Entry')         
a = pd.pivot_table(data=assoc,columns='items',values='Associate')
m = pd.pivot_table(data=mid_senior,columns='items',values='Mid-Senior')
all_ = pd.pivot_table(data=all_levels,columns='items',values='All')
all_data = pd.concat([all_data,e,a,m,all_]).T
all_data.dropna(inplace=True)
```

```python
width=0.2
x = np.arange(len(all_data.index))
y1 = all_data.Entry
y2 = all_data.Associate
y3 = all_data['Mid-Senior']
y4 = all_data.All
plt.figure(figsize=(20,6))
plt.bar(x-0.2, y1, width, color='lightblue')
plt.bar(x, y2, width, color='blue')
plt.bar(x+0.2, y3, width, color='darkblue')
plt.bar(x+0.4, y4, width, color='green')
plt.xticks(x,all_data.index.tolist(),rotation=45)
plt.xlabel("Skill")
plt.ylabel("Percentage")
plt.legend(["Entry", "Associate", "Mid-Senior",'All'])
plt.show()
```
